# 데이터 시각화

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot  as plt
```

```python
# 배열 생성하여 데이터 준비
x = np.arange(1,6)
np.random.seed(3)
y = np.random.randint(1,10,size=5)
print(x)
print(y)
```

```python
# 선그래프 그리기
# plt.plot(x축데이터,y축데이터)
plt.plot(x,y)

# 주피터 노트북 , 구글colab 환경에서는 plot을 그릴때 별도의 코드 없이 그래프가 출력된다
# vscode pytorvh 등 pil.show()까지 작성
plt.show()
```

```python
# 선 그리는 경우 x축 생략 가능함
plt.plot(y)
plt.show()
```

# 선 그래프의 옵션


```python
# 선그래프 스타일 변경
# marker= 포인트에 스타일 변경
# ms= marker 사이즈
# mec = marker edge color = 색깔이니셜사용
# mew = marker...
# mfc = marker...
plt.plot(x,y,ls=":",marker='D',ms=10,mec='r',mew = 3,mfc='g')
plt.show()
```

## 하나의 출력문에 여러개 그래프

```python
plt.plot(x,y,ls = '-', color='b' ,marker='D', ms=10, mec='c',mew=4, mfc='y')
plt.plot(x,y, ls = '--', color='m', lw=5)
plt.plot(x,y, ls = '-.')
plt.plot(x,y, ls = ':')
plt.show()
# 별도의 코드기 없으면 겹쳐서 출력
```

```python
# 하나의 출력문에 여러개 그래프
# plt.subplot()
plt.plot(x,y,ls = '-', color='b' ,marker='D', ms=10, mec='c',mew=4, mfc='y')
plt.plot(x,y, ls = '--', color='m', lw=5)
plt.plot(x,y, ls = '-.')
plt.plot(x,y, ls = ':')
plt.show()
```
