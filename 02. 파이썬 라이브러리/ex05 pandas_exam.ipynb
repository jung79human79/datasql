```python
import numpy as np
```

```python
# 1. np.where 함수 사용하여 평균이 60점 이사이면 pass 아니면 fail 파생변수 만들기
# 파생변수명은 result
exam['result'] = np.where(exam['평균']>=60 , 'pass' ,'fail')
exam
```

```python
#2. apply함수사용 평균이 80이상이면 70이상 60이상 그외는 f
# 파생변수 만들기
# 파생변수명은 grade
def makeGrade(row):
    mean = row['평균']
    if mean>=80:
        return 'A'
    elif mean>=70:
        return 'B'
    elif mean>=60:
        return 'C'
    else :
        return 'F'
```

```python
exam['grade']= exam.apply(makeGrade, axis=1)  # axis=1은 연비평균열의 값을 읽어 등급 설정
exam
```

```python
# 3. nclass별 pass와 fail 학생수 확인하기 --> 오름차순-->nclass,result 컬럼 모두
exam.groupby(['nclass','result']).agg(학생수 = ('result','count'))
```

```python
# 4. nclass별 등급별(a~f) 학생수 확인
exam.groupby(['nclass','grade']).agg(학생수 = ('grade','count'))
```

```python
# 3결과값을 내림차순 (높은 순부터~)
# 번외 3결과값을 fail이 먼저 나옵니다--> 알파펫 오름차순이므로~ --> pass/fail 순으로 바꿔보는 작업
# sort_index 사용
# 시리즈 형식이고 인덱스, 밸류이다
# 밸류가 sort기준이면 by="컬러명" 을 적는데 index기준이면 인덱스는 무조건 시리즈에서 하나이므로 안 적어도 된다
exam.groupby(['nclass','result']).agg(학생수 = ('result','count')).sort_index(ascending=False)
```

```python
# nclass컬럼도 내림차순이 되버리는데 nclass컬럼만 다시 오름차순으로~ 
e1=exam.groupby(['nclass','result']).agg(학생수 = ('result','count')).sort_index(ascending=False)
# 밸류가 sort기준이면 by="컬러명" 을 적는다
e1.sort_values(by='nclass',ascending=True)
```

```python
# 변수 없이 그냥 한 것임-->같은 결과
exam.groupby(['nclass','result']).agg(학생수 = ('result','count')).sort_index(ascending=False).sort_values(by='nclass',ascending=True)
```

```python
# exam 과 teacher프레임 병합
# merge() --> 가로로 병합  --> 컬럼수가 늘어남
# 기준이 되는 컬럼을 잡고 새로운 컬럼으로 생성되어 가로로 병합
# merge(데이터프레임1 ,데이터프레임2, how = 'left or right' ,on='컬럼명' )
pd.merge(exam,teacher,how = 'left',on='nclass')
```

```python
test1 = pd.DataFrame({'id':[1,2,3,4,5],'test':[60,80,70,90,85]})
test2 = pd.DataFrame({'id':[1,2,3,4,5],'test':[70,83,65,95,80]})
display(test1)
display(test2)
```

```python
#concat([데이터프레임1 ,데이터프레임2]) --> 세로로 병합

pd.concat([test1,test2])
```

```python
# 인덱스를 새롭게 부여하여 증가
#pd.concat([],ignore_index =True)
pd.concat([test1,test2],ignore_index =True)
```

## 결측치 와 이상치

## 결측치
- 누락된 값 ,비어있는 값
- 수집 과정에서 오류 발생하거나 포함되어있지 않는 값 가지고 올때
- 데이터의 오류 --> 분석 결과의 왜곡이 발생
- 타이타닉 데이터의 carbin 비어있는 데이터

- 결측치 해결법
    - 1.채워준다  --> 주로 결측치의 갯수가 매우 적을 때 --> 평균값이나 중간값을 채워준다
    - 2. 버린다 -->주로 결측치의 갯수가 너무 많을 때 --> 컬럼자체를 버린다
                --> 결측치가 매우 적어도 버릴 수 있다 --> 결과도출에 큰 영향이 없으므로

###  이상치
- 정상범위에서 크게 벗어난 값
- 논리적으로 존재할 수 없는 값
- 타이타닉에서 나이 데이터 (0~2살:미정확,100살:비현실적인 나이)
- 이상치 해결법
    - 1.바꿔준다  --> 데이터에 따라서 값을 예상 할수 있는 경우 -->바꿔준다
    - 2.버린다  --> 나이가 100살이면 어떻게 바꿔주기 애매하다 --> 버린다
