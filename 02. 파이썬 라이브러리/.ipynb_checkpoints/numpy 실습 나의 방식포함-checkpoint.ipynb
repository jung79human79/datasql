### 최종실습 
- 영화 데이터 평점 분석

```python
import numpy as np
```

```python
# 데이터 불러오기
data = np.genfromtxt('./data/ratings.dat',delimiter='::').astype("int64")
data
```

```python
# user_id , item_id  rating , timestamp(유닉스타임스탬프변환기사용하면시간변경가능)
```

```python
# 배열의 차원 크기,전체요소 갯수 
print('차원: ',data.ndim)
print('크기: ',data.shape)
print('전체요소갯수: ',data.size)
print('타입: ',data.dtype)
```

```python
# 전체 평점 평균 구하기
# 평균--> rating(몇번 열인지 잘 고민)
data[:,2]  # 전체 평점
data[:,2].mean()  # 2:은 안됨
np.mean(data[:,2])
```

```python
# np.unique()--> 데이터의 고유값을 구하는 함수
user_id = np.unique(data[:,0])
print(user_id)
print(user_id.size)
print(user_id.shape)
user_id
```

```python
# 사용자 아이디만 블리언 인덱싱하기
# --> 우선 아이디가 1번인 사람만 추출해보자
data[data[:,0]==1]
```

```python
# 위에서 구한 사용자 아이디가 1인 데이터의 평점 평균 구하기
np.mean(data[data[:,0]==1][:,2])
```

```python
# 사용자별 평균 평점 구하기
#  내 방법
for i in user_id :
    #rs = np.mean(data[data[:,0]==i][:,2])
    rs = data[data[ : , 0 ] == i][:, 2].mean()
    rs2 = np.array([i,rs])
    print(rs2)
```

```python
# 교수님 정답
list1=[]
for i in user_id :
    #rs = np.mean(data[data[:,0]==i][:,2])
    rs = data[data[ : , 0 ] == i][:, 2].mean()
    list1.append(rs)
list1
```

```python
# 1.새로운 교수방법 
list1=[]
for i in user_id :
    #rs = np.mean(data[data[:,0]==i][:,2])
    rs = data[data[ : , 0 ] == i][:, 2].mean()
    list1.append([i,rs])
list1
```

```python
# 2.새로운 교수방법 
result22 = np.array(list1)
result22
```

```python
# 교수님 정답
u_m=np.array(list1)
u_m
```

```python
# 교수님 정답
list2=[]
for i in user_id :
    list2.append([i,list1[i-1]])
list2
```

```python
# 최종 사용자별 평점평균
# 교수님 정답
result = np.array(list2)
result
```

```python
# 블리언 인덱싱 이용해서 평균 평점이 4점 이상인 사용자만 구하기
# 내 방법
for i in user_id :
    #rs = np.mean(data[data[:,0]==i][:,2])
    rs1 = data[data[ : , 0 ] == i][:, 2].mean()
    if rs1 >= 4 :
        rs3 = np.array([i,rs1])
        print(rs3)
        
```

```python
# 블리언 인덱싱 이용해서 평균 평점이 4점 이상인 사용자만 구하기
# 교수님 정답

up4 =result[result[:,1]>=4]
up4
```

```python
print('전체요소갯수:',up4.size)
print('크기:',up4.shape)
print('배열의 길이:',len(up4))
print('하나의열의요소갯수:',up4[:,0].size)
```

```python
# csv파일로 저장하기
np.savetxt('user_rating.csv',result,delimiter=',',fmt = '%.3f')
      
```
