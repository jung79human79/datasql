```python
import pandas as pd
import numpy as np
import matplotlib.pyplot  as plt
```

```python
test1 = pd.DataFrame({'id':[1,2,3,4,5],'min':[60,80,70,90,85]})
test2 = pd.DataFrame({'id':[1,2,3,4,5],'final':[70,83,65,95,80]})
test_fill=pd.concat([test1, test2])
test_fill
```

```python
# fillna()함수
# final컬럼에 nan값을 85.0으로 채운다 
# 동일 컬럼명이므로 덮어쓰기
test_fill['final'] = test_fill['final'].fillna(85.0)
test_fill
```

```python
# final컬럼에 nan값을 45.0으로 채운다 
test_fill['min'] = test_fill['min'].fillna(45.0)
test_fill
```

```python
#dropna() 
# 만약 nan 값이 포함 된 행이 있다면 그행들을 전체 드랍
test_fill.dropna()
```

```python
# [0] -> [행,열]-> 숫자0번째 행만 출력 만약 숫자가 아닌 문자라면 ''로 문자를 기록한다
# iloc는 문자열 안되고 무조건 숫자만 가능
test_fill.loc[0]
```

```python
# [행,열]
# [0~4행,모든 열]
test_fill.iloc[0:5,:]
```
