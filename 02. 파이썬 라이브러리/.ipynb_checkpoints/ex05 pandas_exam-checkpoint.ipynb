```python
#4.데이터를 확인하고 모든 과목의 총점 파생변수 만들기


ss = exam['math']+exam['english']+exam['science']
exam['총점']= ss
exam
```

```python
# 4.데이터를 확인하고 모든 과목의 총점 파생변수 만들기
ss3 = exam[['math','english','science']].sum(axis = 1)
exam['총점2']= ss3
exam
```

```python
#5.모든과목의 평균 파생변수 만들기 (새로운 시리즈만들기)
sss = (exam['math']+exam['english']+exam['science'])/3
exam['평균']=sss
exam
# exam['평균']=exam['총점']/3  
```

```python
#5.모든과목의 평균 파생변수 만들기 (새로운 시리즈만들기)
ss5 = exam[['math','english','science']].mean(axis = 1)
exam['평균2']=ss5
exam
```
