```python
# 1.브라우저 역할을 하는 라이브러리 --> webdriver
from selenium import webdriver as wb
# 2. 컴용 키보드 라이브러리
from selenium.webdriver.common.keys import Keys  #--> 대문자 주의
#3.선택자를 구분지어 주는 라이브러리
from selenium.webdriver.common.by import By   #--> 대문자 주의
import pandas as pd
import time
```

```python
#https://map.naver.com/p/
# 접근법이 어렵다
```

```python
driver = wb.Chrome()
driver.get('https://map.naver.com/p/')
```

```python
# 검색창에 동명동 맛집
# ID는 자바스크립에서 기능단위 실행시 고유값으로 사용 --> 고로 아이디가 계속 변경된다
# 클래스는 여러개 태그에 값을 줄수있다 -> css에서 주로 클래스 사용함
# 그래서 아이디가 아닌 클래스로 선택자를 해보자

# #input_search1708564305717 --> 검색창 태그의 id선택자 --> 근데 페이지를 새로 고침하면 검색창 id선택자가 바뀐다.
# 아이디값으로는 선택자 안됨--> 항상 변하는 원인을 알기 어렵다 --> 네이버 측해서 디도스를 막기위한 보안작업인가?
# while문으로 무한 반복 --> 디도스 공격으로 싸이트가 마비됨 --> 정확한 이유는 모른다.

# 검색창 태그의 아이디선택자가 매번 변경된다
# 변하지 않는 선택자가 필요한다
# 그것이 클래스값이다

# 검색창 태그는 --> input태그 1개 
search = driver.find_element(By.CSS_SELECTOR,'.input_search')
search.send_keys('동명동맛집')

search.send_keys(Keys.ENTER)
# 엔터치면 지도 옆에 새로운 화면이 뜬다
# 그 화면은 다른 url주소(다른 html태그) 즉 iframe태그이다
# 원래 주소 html은 처음화면인 지도화면 페이지이다
# 즉 먼가를 검색하면 나오는 지도는 원본이지만 옆에 나오는 창은 frame태그로 된 다른 주소의 html태그이다
```

```python
# 원본데이터에서 원하는 iframe데이터로 변경해야 한다
# req에서 아이프레임 접근법과는 다르다 --> 접근을 하기 위해서는 원본 주소로 찾아가야 한다 -->ex04 참조

# 내가보는 url이 원본데이터
# 아이프레임의 url은 다른 데이터->별도의 html태그

# 브라우저야~ 바꾸어라~ 'searchIframe'라는 id 선택자를 가진 iframe태그로~ --> 원본데이터에서 원하는 iframe데이터로 변경하는 로직
#()은 매개변수로 세부내용을 표현
driver.switch_to.frame('searchIframe')
# 아래사진 참조 
# 우선 switch_to.frame()해서 브라우저(driver)을 iframe태그로 바꾼 후 거기서 상점명 태그들를(복수) 수집한다.
# 사진을 보면 맨위에 'searchIframe' 이 보이고 맨 아래에 상점명 '마실'이 보인다
# 즉 iframe태그에서 상점명 태그를 수집하는 것이다 --> 상점명이 원본 html태그에 있는 것이 아니므로 frame태그에서 수집하는 것이다.
```

```python
# 가게이름 클릭해서 디테일 정보를 확인하기
# 접근이 불가능한 이유는 아이프레임속의 데이터이기때문

#_pcmap_list_scroll_container > ul > li:nth-child(1) > div.CHC5F > a > div > div > span.TYaxT

title = driver.find_elements(By.CSS_SELECTOR,'#_pcmap_list_scroll_container > ul > li > div.CHC5F > a > div > div > span.TYaxT')
print(len(title))
title

# 중요!! 상점명이 여러개인 경우 내가 스크롤을 수동으로 내린 후 파이썬에서 실행할 때 수집되는 리스트 갯수와
# 그냥 바로 나오는 화면에서 (스크롤을 통해 비동기 통신을 받지 않고=상점 갯수를 더보여주는 행위없이) 
# 파이썬에서 실행할 때 수집되는 리스트 갯수는 후자가 훨씬 적다
```

```python
# 첫번째 가게이름클릭->상세정보
# title (태그가 담긴) 리스트의 첫번째 요소을 클릭해서 화면 띄우고 텍스트를 보자

title[0].click()
#title[0].text #--> 바로 코드쓰면 에러 -->타임슬립줘도 안됨 --> 다음칸에 써야함 
# --> 먼저 클릭 후 실행 한후 text을 봐야함 

```

```python
# title[0]은 text 함수가 적용되는 태그 한개이다
title[0].text
```

```python
# 상세정보의 가게 명을 수집  --> 단수개 driver.find_element
# 상세정보페이지도 또 다른 아이프레임이므로 접근불가

#_title > div > span.Fc1rA
title_detail = driver.find_element(By.CSS_SELECTOR,'#_title > div > span.Fc1rA')
title_detail.text
```

```python
# 상세정보의 가게 명을 수집하기위해 먼저 상세 정보 페이지 즉 새로운 아이프레임에 접근하는 코드가 필요함
# 아이프레임에서 다른 아이프레임으로 이동불가--> 서로모르는 사이이므로
# 해결책은 -> 원본데이터로 이동후 다시 새로운 아이프레임으로 이동해야함

driver.switch_to.default_content()  # 원본페이지는 하나이므로 매개변수 없음 # default가 초기값 즉 원본페이지
driver.switch_to.frame('entryIframe') #새로운 즉 상세 정보 페이지 아이프레임의 아이디

# 프레임 to 프레임은 접근 불가
# 부모원본으로 갔다가 다시 현재 프레임으로 간다
# 당연히 다른 프레임으로 가려면 원본으로 갔다가 가야함
```

```python
# 아이프레임은 독립적인 하나의 파일
# 각각 주소가 다른 프레임 주소을 갖는다
# 부모 자식 -->부모가 원본 --> 1번프레임과 2번프레임은 자식
# 근데 자식끼리는 연관 없음
```

```python
title_list = []
try:
    for i in tqdm(range(10)): 
    #driver.switch_to.default_content()
        time.sleep(1)
        driver.switch_to.frame('searchIframe') # 아이프레임의 아이디
        time.sleep(1)
        title = driver.find_elements(By.CSS_SELECTOR,'#_pcmap_list_scroll_container > ul > li > div.CHC5F > a > div > div > span.TYaxT')
        title[i].click()
        time.sleep(1)
        driver.switch_to.default_content()
        time.sleep(1)
        driver.switch_to.frame('entryIframe')
        time.sleep(1)
        title_detail = driver.find_element(By.CSS_SELECTOR,'#_title > div > span.Fc1rA')
        title_list.append(title_detail.text)
        driver.switch_to.default_content()
        time.sleep(1)

except:
    print('더보기왼료')
```

```python
for i in tqdm(range(10)): 
    print("start")
    driver.switch_to.frame('searchIframe') # 아이프레임의 아이디
    time.sleep(1)
    title = driver.find_elements(By.CSS_SELECTOR,'#_pcmap_list_scroll_container > ul > li > div.CHC5F > a > div > div > span.TYaxT')
    title[i].click()
    time.sleep(1)
    print(1)
    driver.switch_to.default_content()
    driver.switch_to.frame('entryIframe')
    time.sleep(1)
    print(2)
    title_detail = driver.find_element(By.CSS_SELECTOR,'#_title > div > span.Fc1rA')
    title_list.append(title_detail.text)
    driver.switch_to.default_content()
    time.sleep(1)
    print(3)
```
