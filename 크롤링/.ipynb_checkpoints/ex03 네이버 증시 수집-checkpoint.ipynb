```python
import requests as req
from bs4 import BeautifulSoup as bs
```

```python
# 네이버 증시 수집 https://finance.naver.com/sise/sise_market_sum.naver
```

```python
# 1. req로 페이지 정보 요청
res=req.get('https://finance.naver.com/sise/sise_market_sum.naver')
res.text
```

```python
# 2. bs을 통해서 soup 데이터로 변형
soup = bs(res.text,"lxml")
soup
```

```python
# 3.종목명을 수집
title = soup.select("a.tltle")
print(len(title))
title
```

```python
# 4. 가격수집
# 선택자를 한번에 복사해 오는 방법
# 개발지도구-> 원하는 요소-> 우클릭-> copy->copyselector
# 주의점! 반드시 내가 선택한 요소 한개의 선택자만 반환되는 단점

# 선택자중 id,가상클래스(:nth-child(n))는 단수개를 의미하는 선택자
# 단수개를 삭제하자 --> 그래야 가격정보 50개를 수집할 수 있다
# 선택자 수정은 반드시 뒤에서 검색하면서 살피자
# tr태그를 지우자 --> 2행의 3열이 73500원 -->tr:nth-child(2) > td:nth-child(3)
# 어차피 열은 가져오고 행을 지워야 하므로 tr태그를 지우자
# tr:nth-child(2)에서 :nth-child(2)만 즉 가상클래스를 삭제

# 복수개를 수집할 때는 반드시 선택자를 수정
# 공식 --> 단수개를 의미하는 선택자를 제거(id,:nth-child(n))

#soup.select("#contentarea > div.box_type_l > table.type_2 > tbody > tr:nth-child(2) > td:nth-child(3)")
price = soup.select("#contentarea > div.box_type_l > table.type_2 > tbody > tr >td:nth-child(3)")
price
```
