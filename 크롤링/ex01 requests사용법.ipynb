# requests사용법
- 브라우저 대신 파이썬에서 페이지 정보를 요청 할 때 사용

```python
import requests as req
```

```python
# 1.네이버에게 메인 페이지를 요청
# req를 통해서 네이버 페이지 가져와~
req.get('http://www.naver.com')
res = req.get('http://www.naver.com')
# 응답은 res는 딕셔너리 타입으로 반환.
# Response[200,300] --> 통신에 성공했습니다
# <Response [200]>  --> 200은 응답코드
# [400] --> 통신에 실패(클라이언트의 요청의 문제) --> 수정가능
# [500] --> 통신에 실패(서버에 문제가 있다) --> 수정 불가
res
```

```python
# 응답코드안에 글자 정보만 보여줘
# 리스트 = 배열은  list[0] ,arr[0] 이렇게 밸류를 추츨
# 딕셔너리 =  딕{'키':'밸류'} -> 딕['키'] 이렇게 밸류를 추츨
# 여기서는 딕셔너리 = 객체라고 정의하고  dic.k(키) --> res.text
# 응답받은 데이터에서 html내용만 뽑을 때는 res.text
res.text
```

# 멜론사이트의 정보를 요청

```python
req.get('http://www.melon.com/chart/index.htm')
# 멜론에서 브라우저가 아닌 파이썬에서 요청하는 것을 알고 서버에서
# 차단해버림 --> 브라우저인 척 우회한다
# <Response [406]> --> 클라이언트의 요청의 문제
# 응답코드가 406 --> 요청하는 클라이언트의 문제
# 멜론서버가 요청하는 주체가 브라우저 아님을 인지
# 브라우저인척 속여서 요청을 보내자
# 내가 원하는 사이트가 응답이 406이 나오면 이 방법을 무조건 시도해보자

# res = req.get('http://www.melon.com/chart/index.htm')
```

```python
res.text
```

```python
# 브라우저인척 속이기 위해서 head정보를 제작
# 개발자도구에서 네트워크에서 index.htm에서 headers에서 맨 아래에서 복사따옴
# 아래 코드는 모든 사이트에서 왠만하면 공통요소이다.
head={"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/121.0.0.0 Safari/537.36"}
```

```python
# 매개변수에 headers=head정보를 담아서 요청
# headers는 고정값이다 --> 개발자도구
req.get('https://www.melon.com/chart/index.htm',headers = head )
res=req.get('https://www.melon.com/chart/index.htm',headers = head )
```

```python
res.text
```

# beatifulSoup 라이브러리
- req로 받아온 텍스트 데이터를 컴이 이해 할수 있게 html로 변환 시켜주는 라이브러리

```python
# 임폴트 할시 stone사용 안함
# bs4폴더로부터 임폴트한다.
from bs4 import BeautifulSoup as bs
```

```python
# bs을 이용해서 req의 텍스트 정보를 html로 변환
# 문자열이 아니다 따옴표 없다 --> 즉 컴이 이해하는 htm코드 -->즉 컴에 명령해서 특정 데이터를 수집할수 있게된다.
# 컴은 태그로만 제목이나 문단인지를 알수 있다
# 그것을 우리는 명령한다

soup = bs(res.text,"lxml")
```

```python
# soup 데이터에서 컴에게 데이터 선택해! 필요한 태그들만!
# 모든 a 태그를 가져온다 
# soup데이터에서 필요한 데이터를 추출 ->soup.select()
# select함수는 리턴 데이터가 리스트로 넘어온다 --> 인덱스,반복문 활용해야함
# select_one --> 한개만 리턴 --> 리스트의 첫번째만 리턴 --> 사용 잘 안함
title = soup.select("a.news_tit")
title
# 여기서는 클래스 선택자이므로 .찍는다 news_tit은 클래스 이름
# a태그 중 선택자를 알아야 한다 --> 개발자도구 사용 -> 
# 원하는 부분을 우 클릭후 검사를 누르면 해당 태그를 볼수 있다
```

```python
# 글자정보만 출력 .text를 사용
# a태그에 1번 인덱스 text 가져오기
title[1].text 
```

```python
# 모든 글자 정보만 출력
# 
for i in range(len(title)) :
    print(title[i].text)
```

```python
for i in title :
    print(i.text)
```

### req를 활용한 데이터 수집 프로세스
1) 데이터 요청 -> req를 활용
2) 응답코드를 확인 -> 200
3) bs을 통해서 텍스트 -> html 변형 = 컴이 이해할수 있게 만들기 위해서
4) 변형된 데이터중이 내가 필요한 요소 수집 -> soup.select('선택자')
5) html데이터 중에 필요한 텍스트만 추출 -> 사용자가 사용하기 위해서
6) 정제된 데이터를 활용
