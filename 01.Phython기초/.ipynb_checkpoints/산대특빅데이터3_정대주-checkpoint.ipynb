### 3.아래와 같이 리스트를 만들고 요소의 짝수 홀수를 구분하여 개수를 세는 프로그램

```python
list = [1,50 ,27, 19,37,33,5,3,48]
a=0
b=0
for i in range(len(list)) :
    if list[i]%2 == 0 :
        a+=1
    else :
        b+=1
print(f'짝수{a}개 ,홀수{b}개')
    
```

### 4.다음 에 들어있는 데이터를 이용해 아래와 같이 성적표를 만드는 score_list 프로그램을 작성하시오

```python
score_list = [['김미희',89,78,99],
              ['서대희',67,88,77],
              ['강예진',100,55,96]]

print("=============================================")
print("이름\tPYTHON\tJAVA\tDB\t평균\t학점")
print("=============================================")
av=0
gr=''
for a,b,c,d in score_list : 
    av = int((b+c+d)/(len(score_list[0])-1))
    if av >=90 :
        gr='A'
    elif av >=80 :
        gr='B'
    elif av >=70 :
        gr='C'
    elif av >=60 :
        gr='D'
    else  :
        gr='F'
    
    print(f'{a}\t{b}\t{c}\t{d}\t{av}\t{gr}')
print("=============================================")
bb=0
cc=0
dd=0
for a,b,c,d in score_list : 
    bb += b
    cc += c
    dd += d
    av2 =int(bb/len(score_list))
    av3 =int(cc/len(score_list))
    av4 =int(dd/len(score_list))
print(f'평균\t{av2}\t{av3}\t{av4}')
```

### 5. X, Y1, Y2 [ ] 리스트 자료를 활용하여 실행결과 와 같은 그래프를 출력하는 프로그램을 작성하시오


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot  as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
x = [2,3,4,5,6,7,8]
y1 = [5,7,3,9,1,8,6]
y2 = [9,5,7,2,3,7,1]
plt.plot(x,y1,label = 'y1',c='r')
plt.plot(x,y2,label = 'y2',c='b')
plt.grid()
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

###  6. Traffic_Accident_2017.csv ‘ ’ 데이터를 이용하여 차대 사람 인 사건 중 야간에 발생한 사건에 대해 아래 주어진 조건을 지켜 다음과 같이 시각화 하시오.



```python
data = pd.read_csv('./Traffic_Accident_2017.csv', encoding = 'euc-kr')
data
```

```python
data.columns
```

```python
data['사고유형_대분류'].unique()
```

```python
data['주야'].unique()
```

```python
# 1. 조건 사고유형 대분류가 차대사람
ctc = data[data['사고유형_대분류'] == '차대사람']
ctc
```

```python
#  2.조건 발생시간대가 야간 인 경우

ctc1 = ctc[ctc['주야'] == '야간']
ctc1
```

```python
# 3.조건--> 사고유형 대분류가 차대사람 이면서 발생시간대가 야간 인 경우의 사고 건수 카운트 -->교집합
temp=ctc1['요일'].value_counts()
temp
```

```python
# 3.인덱스 순서를 조절하고 y축에 저장
temp=ctc1['요일'].value_counts()
temp

# 다중인덱싱
y=temp[['월','화','수','목','금','토','일']]

print(y.values)
y
```

```python
# x축저장
x = y.index
```

```python
# 막대그래프 생성
plt.figure(figsize=(15,8))
plt.bar(x,y)
plt.xlabel('요일')
plt.ylabel('사고건수')
plt.title('요일별 야간 차대사람 교통사고 건수')
plt.ylim(0,200)
plt.show()
```

```python
# 다른 방법 --> 선그래프 만들고 그 다음 막대 그래프 만들기
```

```python
# 선그래프를 위해 데이터프레임으로 생성
ctcp1=ctc1[['사상자수','요일']].groupby('요일').count()
ctcp1
#ctcp1.index=['월','화','수','목','금','토','일']    #정상수치아님--> 인덱스 이름만 변경될뿐 값은 변경안됨
#unstack( )을 이용해서, Series형 데이터를 DataFrame형의 데이터로 변경할 수 도 있다.
#stack( )을 이용하면, df을 Series로 변경한다.

```

```python
#stack( )을 이용하면, df을 Series로 변경한다.
aaa = ctcp1.stack( )
aaa
```

```python
#stack( )을 이용하면, df을 Series로 변경한 후 --> 다중인덱싱으로 순서를 변경한다.
# depth가 2이다 인덱스=['월','화','수','목','금','토','일'] 값=[123,........]
aaa = aaa[['월','화','수','목','금','토','일']]
aaa
```

```python
#unstack( )을 이용해서, Series형 데이터를 DataFrame형의 데이터로 변경할 수 도 있다.-->근데 다시 정렬 무너진다.
bbb = aaa.unstack( )
bbb
```

```python
# reindex(index=[]) --> 인덱스가 정렬이 되고 값도 함께 가져온다
ccc = bbb.reindex(index=['월','화','수','목','금','토','일'])
ccc
```

```python
# 시리즈를 추가해서 해보자~~
```

```python
# 긱 축 저장
x = ctcp1.index
y = ctcp1.values
print(x)
print(y)
print(x.shape)
print(y.shape)
# x축은 리스트 y축은 배열
```

```python
# 선그래프 생성
plt.figure(figsize=(15,8))
plt.plot(x,y)
plt.xlabel('요일')
plt.ylabel('사고건수',rotation = 0)
plt.title('요일별 야간 차대사람 교통사고 건수')
plt.show()
```

```python
# 막대그래프를 위해 y축을 1차원 배열(시리즈)로 변경
# 데이터 프레임의 인덱스를 정렬하기 --> 가장 큰 궁금점!!

print(len(y)) #-->7
y_bar = []
for i in range(len(y)) :
    y_bar.append(y[i][0])


plt.figure(figsize=(15,8))
plt.bar(x,y_bar)
plt.xlabel('요일')
plt.ylabel('사고건수')
plt.title('요일별 야간 차대사람 교통사고 건수')
plt.ylim(0,200)
plt.show()
```
